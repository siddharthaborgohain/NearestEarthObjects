There are infinite number of objects in the outer space. Some of them are closer than we think. Even though we might think that a distance of 70,000 Km can not potentially harm us, but at an astronomical scale, this is a very small distance and can disrupt many natural phenomena. These objects/asteroids can thus prove to be harmful. Hence, it is wise to know what is surrounding us and what can harm us amongst those. Thus, this dataset compiles the list of NASA certified asteroids that are classified as the nearest earth object.

We have collected a dataset from https://www.kaggle.com/datasets/sameepvani/nasa-nearest-earth-objects?select=neo.csv .

## Feature Engineering

Estimated Diameter difference in km('est_diameter_diff') is the difference between Maximum Estimated Diameter in Kilometres('est_diameter_max') and Minimum Estimated Diameter in Kilometres('est_diameter_min')
so  the equation is 
                                                 
                                                      'est_diameter_diff'=['est_diameter_max'-'est_diameter_min']
    I have taken this parameter as orbital deviation can increase the chance of an NEO  to be more hazardous because if the orbit changes and reached very near to earth then the NEO object will be pulled by earth,s gravitation .

```python

fe=df['est_diameter_max']-df['est_diameter_min']
```

```python
df.insert(4,'est_diameter_diff',fe)
```

```python
df.head()
```

```python
df['name']=df['name'].apply(lambda x: x.split("(")[1])
```

## Model

```python
df.head()
```

```python
X=df.iloc[:,1:-1]
X.head()
```

```python
y=df.iloc[:,-1]
y.head()
```

```python
y.value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE()
```

```python
from collections import Counter
print("before smote",Counter(y))
```

```python
X,y=smote.fit_resample(X,y)
```

```python
print("after smote",Counter(y))
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X=scaler.fit_transform(X)
```

```python
# now finding the top 5 important features
from sklearn.ensemble import ExtraTreesClassifier
model=ExtraTreesClassifier()
model.fit(X,y)
```

```python
model.feature_importances_
```

```python
### plot a graph of feature inportance
feature_importance=pd.Series(model.feature_importances_,index=['est_diameter_min','est_diameter_max','relative_velocity','est_diameter_diff','miss_distance','absolute_magnitude'])
feature_importance.nlargest(5).plot(kind='barh')
plt.show()
```

```python
from sklearn.model_selection import StratifiedKFold
folds = StratifiedKFold(n_splits=10)
```

```python
for i,j in folds.split(X,y):
    X_train, X_test, y_train, y_test = X[i],X[j],y[i],y[j]
```

```python
print(X_train.shape,y_train.shape)
```

```python
 print(X_test.shape,y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
```

```python
rf=RandomForestClassifier()
```

```python
rf_random=RandomizedSearchCV(rf,param_distributions={'criterion':["gini"]},cv=folds,verbose=2)
```

```python
rf_random.fit(X_train,y_train)
```

```python
rf_random.best_score_
```

```python
y_pred=rf_random.predict(X_test)
y_pred
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,y_pred))
```

```python

confusion_matrix(y_test,y_pred)
```

```python
sns.heatmap(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.metrics import plot_roc_curve
plot_roc_curve(rf_random,X_test,y_test)
```
